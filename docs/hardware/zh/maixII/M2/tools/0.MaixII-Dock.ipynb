## 准备工作

1. 选择一条可以传输数据的 Tpye-C 数据线（如购买时附带的数据线）。
3. 插入带有系统的 SD 卡，可在购买时选择预烧录套餐（没有的需要[自行烧录](https://wiki.sipeed.com/hardware/zh/maixII/M2/flash.html)）。
2. 将 板子 与 电脑 通过 OTG USB 口连接，确认设备通电亮起电源红灯。

    

4. 直到屏幕显示如下内容，并且会弹出一个 U 盘，这意味着板子的系统已经准备就绪。（[没有U盘弹出](/soft/maixpy3/zh/question/Maixpy3_IDE.html#没有U盘设备弹出)）





## 安装驱动

1. 安装并启动 [MaixPy3 IDE](https://wiki.sipeed.com/soft/maixpy3/zh/tools/MaixPy3_IDE.html)，在任务栏中的图标中找到 或 ，在图标的右键菜单中选择 PhoenixInstall -> 启动，开始进行驱动安装。



2. 按 `下一步` 直到安装成功即可。





## 开发入门

建议小白基于 MaixPy3 IDE 使用 Python 进行开发，具体的使用方法查看 MaixPy3 的[使用文档](/soft/maixpy3/zh/index.html)

## 如何运行代码

使用 MaixPy3 IDE 或者在命令行运行代码。（使用命令行需要使用其他[连接方式](#其他连接方式教程)）

### 关于 IDE 

MaixPy3 IDE 是通过对 [jupyter](https://www.baidu.com/s?ie=UTF-8&wd=jupyter) 进行魔改形成，会在本机的浏览器中启动。你可以在这里进行代码的分步调试，保留报错信息。当右角显示为 `RPyc-Python` 时，IDE 所调用的是开发板内部的 Python，这里运行代码和在开发板的终端上运行代码所输出的结果是一致的。

### MaixPy3 IDE 基本使用 

在下载站中获得 [MaixPy3 IDE](https://dl.sipeed.com/shareURL/MaixII/MaixPy3-IDE) 安装，双击打开并安装，安装结束之后，会以网页的形式在电脑默认浏览器中打开。

> 下载站下载会比较慢，推荐百度云下载：链接：<https://pan.baidu.com/s/1C7Jl3rXticeJsCgH-fIcbQ> 提取码：2333

#### 如何运行代码
点击选择代码块，点击上方工具中的运行，即可运行代码并输出结果。



#### 如何停止运行代码

选择正在运行的代码，点击上方工具栏中的停止，即可停止运行代码。



![](./asserts/Maixpy3IDE.gif)


#### 注意事项

IDE 的每个代码单元块是独立的，只有定义为全局变量才能在不同的代码单元块中传递。运行也是按开始的顺序进行排队运行，当前一个代码没有运行结束的时候，后一个代码是不会运行。

当出现代码运行之后没有出现结果，或者是卡住了，重启 IDE 和重启开发板就可以解决了。

### 运行代码

按照上面运行代码的教程，在 IDE 中运行下列代码，测试板子是否连接正常

```python
import platform
print(platform.platform())

import time
print(time.asctime())
```

上述结果可知：

一、运行这段代码的时间是 \[ rpyc-kernel \]( running at Mon Dec  6 17:29:31 2021 )

二、运行这段代码的平台是 Linux-4.9.118-armv7l-with-libc

三、运行这段代码的时候，板子系统时间是 Tue Dec  7 01:56:33 2021。

当代码运行的时间为当前时间，并打印出以上代码，说明开发板已经连接上并可以正常的使用





## 连接网络

开发板上的 OTG 接口与电脑连接之后，就会在资源管理器中得到一个 U盘 设备。通过记事本打开里面名为 `wpa_supplicant.conf` 文件




![](./asserts/net.gif)

在双引号中输入WiFi账号和密码。(只支持连接2.4G频段的WIFI，每个人的网络环境不同，也有可能连接不上)

## 更新 MaixPy3

手动下载最新的 MaixPy3 [安装包](https://pypi.org/project/maixpy3/#history)

> 安装包的网页打开会比较慢



下载带有 cp8 的安装包，cp9 是给别的平台使用的。将这个安装的名字修改成 `maixpy3-9.9.9-cp38-cp38-linux_armv7l.whl`,直接存放到开发板中，重启开发板就会自动更新和安装 MaixPy3。




## 开机自动运行脚本

将需要开机自动运行的脚本文件名修改成 `main.py` 替换到
开发板中的 U盘 中即可。

> 存放之前要注意，确保脚本是可运行的代码，脚本运行出错只会中断运行，不会有信息提示出现。

> u 盘位置对应着开发板中的 root 文件夹
